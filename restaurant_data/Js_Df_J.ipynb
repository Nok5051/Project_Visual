```python
import json
import requests
import pandas as pd
import numpy as np
```

```python
key = "76466b686e6a736e3132305a486c4d4c"

startpage = "1"
endpage = "1000"

url="http://openapi.seoul.go.kr:8088/" + key + "/json/LOCALDATA_072404/"+ startpage + "/" + endpage + "/"

resp = requests.get(url)
# print(resp)

r_dict = json.loads(resp.text)
# print(type(r_dict))
# print(r_dict)

doc = r_dict["LOCALDATA_072404"]["row"]
# print(doc)

Data = pd.DataFrame(doc)
# print(Data)
```

```python
for i in np.arange(1000, 480000, 1000):
    
    try:
        a = 1 + i
        b = 1000 + i

        startpage = str(a)
        endpage = str(b)

        url="http://openapi.seoul.go.kr:8088/" + key + "/json/LOCALDATA_072404/"+ startpage + "/" + endpage + "/"

        resp = requests.get(url)

        r_dict = json.loads(resp.text)

        doc = r_dict["LOCALDATA_072404"]["row"]

        Data_1 = pd.DataFrame(doc)

        Data = pd.concat([Data, Data_1], axis=0, ignore_index=False)
    
    except:
        pass
```
